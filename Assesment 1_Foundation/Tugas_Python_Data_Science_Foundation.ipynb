NO.3 Buatlah program untuk menghitung BMI

```python
def determine_bmi_status(row):
    bmi_value = row['BMI']

    if bmi_value < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi_value <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi_value <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi_value <= 34.9:
        return 'Obesity'

df['Status BMI'] = df.apply(lambda row: determine_bmi_status(row), axis=1)
```

# Soal 2
Perusahaan Invesitasi Jamin-Sugih CO kami memberikan penawaran invesitasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun

---
Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk :

---

1. Modal invesitasi (bilangan bulat antara 0 - 1000.000.000)
2. Jangka Waktu invesitasi (bilangan bulat antara 1-25 dalam tahun)


---
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan


---
Program akan menampilkan informasi akumulasi invesitasi setiap tahun menggunakan iterasi untuk setiap tahun


---
contoh :

Input

Money Capita : 100000

Investment Tenure (years) : 2

Output

Year 1, your total investment : 125000.0

Year 2, yout total investment : 156250.0






```python
#fungsi program
def investasi(capital, tenure):
    if (capital <= 0 or capital > 1000000000) or (tenure < 1 or tenure > 25):
        print("Error: Modal atau tenure tidak sesuai ketentuan.")
        return

    keuntungan_tahunan = 0.25
    total_investasi = capital

    for year in range(1, tenure + 1):
        total_investasi +=  total_investasi * keuntungan_tahunan
        print(f"Year {year}, your total investment : {total_investasi:.1f}")


# Input dari pengguna
try:
    capital = int(input("Money Capital : "))
    tenure = int(input("Investment Tenure (years) : "))
    investasi(capital, tenure)
except:
    print("Error: masukan angka dalam bentuk bilangan bulat.")

```
